# Ground Truth Calculations for Baseline Questions

This section calculates all ground truth values for questions in `baseline_describe_data.csv`.

## Housing Dataset Statistics

```python
# Q1 & Q13: Average house price (mean and median)
price_mean = housing['price'].mean()
price_median = housing['price'].median()
price_std = housing['price'].std()

print("=" * 80)
print("Q1 & Q13: Average House Price")
print("=" * 80)
print(f"Mean: ${price_mean:,.2f}")
print(f"Median: ${price_median:,.2f}")
print(f"Std Dev: ${price_std:,.2f}")
print(f"\nGround Truth: \"Mean: ${price_mean:,.2f} (Median: ${price_median:,.2f}, Std Dev: ${price_std:,.2f})\"")
```

```python
# Q2: Rows and columns in housing dataset
print("=" * 80)
print("Q2: Dataset Dimensions")
print("=" * 80)
print(f"Rows: {housing.shape[0]}")
print(f"Columns: {housing.shape[1]}")
print(f"\nGround Truth: \"{housing.shape[0]} rows, {housing.shape[1]} columns\"")
```

```python
# Q3: Price range
price_min = housing['price'].min()
price_max = housing['price'].max()
price_range = price_max - price_min

print("=" * 80)
print("Q3: Price Range")
print("=" * 80)
print(f"Min: ${price_min:,.2f}")
print(f"Max: ${price_max:,.2f}")
print(f"Range: ${price_range:,.2f}")
print(f"\nGround Truth: \"Min: ${price_min:,.2f}, Max: ${price_max:,.2f}, Range: ${price_range:,.2f}\"")
```

```python
# Q4 & Q18: Bedroom distribution
bedroom_counts = housing['bedrooms'].value_counts().sort_index()

print("=" * 80)
print("Q4 & Q18: Bedroom Distribution")
print("=" * 80)
print(bedroom_counts)

# Format for ground truth
bedroom_str = ", ".join([f"{int(bedrooms)} bedroom{'s' if bedrooms != 1 else ''}: {count}" 
                          for bedrooms, count in bedroom_counts.items()])
print(f"\nGround Truth Q4: \"{bedroom_str}\"")

# Q18: Percentage of 3 bedroom houses
three_bedroom_count = bedroom_counts.get(3, 0)
three_bedroom_pct = (three_bedroom_count / len(housing)) * 100
print(f"\nGround Truth Q18: \"{three_bedroom_pct:.2f}% ({three_bedroom_count} out of {len(housing)} houses)\"")
```

```python
# Q5 & Q11: Key statistics summary
print("=" * 80)
print("Q5 & Q11: Key Statistics Summary")
print("=" * 80)
print(housing.describe())

area_mean = housing['area'].mean()
area_median = housing['area'].median()

print(f"\nGround Truth Q5: \"Price - Mean: ${price_mean:,.2f}, Median: ${price_median:,.2f}, Std: ${price_std:,.2f}; Area, bedrooms, bathrooms, stories statistics\"")
print(f"\nGround Truth Q11: \"Price mean: ${price_mean/1e6:.2f}M, Area, Bedrooms (2-3 most common), Bathrooms, Stories - all numerical summaries\"")
```

```python
# Q16: Standard deviation of house prices
print("=" * 80)
print("Q16: Standard Deviation of House Prices")
print("=" * 80)
print(f"Std Dev: ${price_std:,.2f}")
print(f"\nGround Truth: \"${price_std:,.2f} (indicates significant price variability)\"")
```

```python
# Q19: Correlation between house size and price
correlation = housing['area'].corr(housing['price'])

print("=" * 80)
print("Q19: Correlation Between Area and Price")
print("=" * 80)
print(f"Pearson correlation: {correlation:.4f}")

if correlation > 0.7:
    strength = "strong positive"
elif correlation > 0.4:
    strength = "moderate positive"
elif correlation > 0.2:
    strength = "weak positive"
else:
    strength = "very weak or no"

print(f"Interpretation: {strength} correlation")
print(f"\nGround Truth: \"{correlation:.4f} ({strength} correlation between area and price)\"")
```

```python
# Missing values in housing dataset
print("=" * 80)
print("Missing Values - Housing")
print("=" * 80)
missing_housing = housing.isnull().sum()
print(missing_housing)

if missing_housing.sum() == 0:
    print(f"\nNo missing values in housing dataset")
else:
    print(f"\nTotal missing values: {missing_housing.sum()}")
```

## Coffee Shop Sales Statistics

```python
# Calculate total sales amount
coffee['total_sales'] = coffee['transaction_qty'] * coffee['unit_price']
```

```python
# Q6: Number of coffee shop sales records
print("=" * 80)
print("Q6: Number of Coffee Shop Sales Records")
print("=" * 80)
print(f"Total records: {len(coffee):,}")
print(f"\nGround Truth: \"{len(coffee):,} records\"")
```

```python
# Q7, Q14, Q20: Coffee sales statistics
sales_mean = coffee['total_sales'].mean()
sales_median = coffee['total_sales'].median()
sales_std = coffee['total_sales'].std()
sales_min = coffee['total_sales'].min()
sales_max = coffee['total_sales'].max()

print("=" * 80)
print("Q7, Q14, Q20: Coffee Sales Statistics")
print("=" * 80)
print(f"Mean: ${sales_mean:.2f}")
print(f"Median: ${sales_median:.2f}")
print(f"Std Dev: ${sales_std:.2f}")
print(f"Min: ${sales_min:.2f}")
print(f"Max: ${sales_max:.2f}")
print(f"Range: ${sales_min:.2f}-${sales_max:.2f}")

print(f"\nGround Truth Q7: \"Mean: ${sales_mean:.2f} (Median: ${sales_median:.2f}, Range: ${sales_min:.2f}-${sales_max:.2f})\"")
print(f"\nGround Truth Q14: \"Mean: ${sales_mean:.2f}, Median: ${sales_median:.2f}, Std Dev: ${sales_std:.2f}, Min: ${sales_min:.2f}, Max: ${sales_max:.2f}\"")
print(f"\nGround Truth Q20: \"Min: ${sales_min:.2f}, Max: ${sales_max:.2f}\"")
```

```python
# Q8: Date range
coffee['transaction_date'] = pd.to_datetime(coffee['transaction_date'])
min_date = coffee['transaction_date'].min()
max_date = coffee['transaction_date'].max()
date_range_days = (max_date - min_date).days
date_range_months = date_range_days / 30

print("=" * 80)
print("Q8: Date Range")
print("=" * 80)
print(f"Start: {min_date.date()}")
print(f"End: {max_date.date()}")
print(f"Duration: {date_range_days} days (~{date_range_months:.1f} months)")
print(f"\nGround Truth: \"{min_date.date()} to {max_date.date()} ({date_range_months:.0f} months)\"")
```

```python
# Q9 & Q10: Most common products and distribution
product_counts = coffee['product_category'].value_counts()

print("=" * 80)
print("Q9 & Q10: Product Distribution")
print("=" * 80)
print(product_counts)

# Q9: Top products with counts
top_products = []
for product, count in product_counts.head().items():
    top_products.append(f"{product}: {count:,}")
products_str = ", ".join(top_products)
print(f"\nGround Truth Q9: \"{products_str}\"")

# Q10: Distribution with percentages
dist_products = []
for product, count in product_counts.items():
    percentage = (count / len(coffee)) * 100
    dist_products.append(f"{product} ({percentage:.1f}%)")
dist_str = ", ".join(dist_products)
print(f"\nGround Truth Q10: \"{dist_str}\"")
```

```python
# Q13: Unique product types
unique_products = coffee['product_category'].nunique()

print("=" * 80)
print("Q13: Unique Product Types")
print("=" * 80)
print(f"Unique categories: {unique_products}")
print(f"\nGround Truth: \"{unique_products} unique product categories\"")
```

```python
# Q17: Day of week with most sales
coffee['day_of_week'] = coffee['transaction_date'].dt.day_name()
day_counts = coffee['day_of_week'].value_counts()

print("=" * 80)
print("Q17: Sales by Day of Week")
print("=" * 80)
print(day_counts)

top_day = day_counts.index[0]
top_day_count = day_counts.iloc[0]

# Build ground truth string with available days
if len(day_counts) >= 3:
    print(f"\nGround Truth: \"{top_day} with {top_day_count:,} sales (followed by {day_counts.index[1]}: {day_counts.iloc[1]:,}, {day_counts.index[2]}: {day_counts.iloc[2]:,})\"")
elif len(day_counts) == 2:
    print(f"\nGround Truth: \"{top_day} with {top_day_count:,} sales (followed by {day_counts.index[1]}: {day_counts.iloc[1]:,})\"")
else:
    print(f"\nGround Truth: \"{top_day} with {top_day_count:,} sales\"")
```

```python
# Q21: Missing values in both datasets
print("=" * 80)
print("Q21: Missing Values - Coffee")
print("=" * 80)
missing_coffee = coffee.isnull().sum()
print(missing_coffee)

if missing_coffee.sum() == 0:
    print(f"\nNo missing values in coffee dataset")
else:
    print(f"\nTotal missing values: {missing_coffee.sum()}")

# Combined ground truth for Q21
housing_missing_total = housing.isnull().sum().sum()
coffee_missing_total = missing_coffee.sum()
print(f"\nGround Truth Q21: \"Housing: {housing_missing_total} missing values, Coffee: {coffee_missing_total} missing values\"")
```
